```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

$\mu$ = $M_2$/{$M_1$ + $M_2$}   
$M_1$ + $M_2$ = 1$M_{sun}$  
$a_{bin}$ = 1 au   
G = 1  
$T_{p}$ = 2$\pi\sqrt[2]{a_{p}}^{3}$  
$v_{esc}$ = $\sqrt[2]\frac{2}{a_p}$   
$m_{planet}$ = $3*10^{-6}$$M_{sun}$

```python
def setupSimulation(u, e, a_p, inc):
    #Initial Condition
    m1 = 1-u
    m2 = u
    m3 = 3e-6
    p=np.pi/2
    sim = rebound.Simulation()
    sim.add(m=m1)#Star A
    sim.add(m=m2, a=1., e=e)#Star B
    sim.add(m=m3, a=a_p, inc=inc) #Planet ABc
    sim.move_to_com()
    fig = rebound.OrbitPlot(sim,slices=0.8,xlim=[-5.,5],ylim=[-5.,5])# this is the initial state of orbit
    return sim

def aHW99(e, u):
    a = 1.6 + 5.1*e 
    -2.22*(e*e) + 4.12*u 
    -4.27*e*u - 5.09*(u*u) 
    +4.61*(e*e)*(u*u)
    return a
```

```python
#Initial Condition
e = 0.4
u = 0.1
a_p = 0.93*aHW99(e, u)
inc0 = 0.
inc1 = np.pi/3
inc2 = 2*inc1
inc3 = 3*inc1
v_esc2 = 2/a_p # the square of v_escape
t = 2*np.pi*a_p*np.sqrt(a_p)
```

```python
t
```

```python
#inclination=0
#Simulation 
sim0 = setupSimulation(u, e, a_p, inc0)
Noutputs = 1000*100
times = np.linspace(0, 1000*t, Noutputs)
distances = np.zeros(Noutputs)
velocity2 = np.zeros(Noutputs)
ps0 = sim0.particles # ps is now an array of pointers. It will update as the simulation ru
result = 0
for i, time in enumerate(times):
    sim0.integrate(time)
    dp = ps0[1] - ps0[2]
    dv = ps0[2]
    distances[i] = np.sqrt(dp.x*dp.x+dp.y*dp.y+dp.z*dp.z)
    velocity2[i] =  (dv.vx*dv.vx+dv.vy*dv.vy+dv.vz*dv.vz)
if max(velocity2) > v_esc2 or min(distances) < 1.:
    print('unstable')
else:
    print('stable')

#y0=velocity2[:]
```
