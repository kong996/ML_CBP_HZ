```python
import glob, os
import pandas as pd
```

```python
path = './'
file = glob.glob(os.path.join(path, "earth_train**.csv"))
dl = []
for f in file:
    dl.append(pd.read_csv(f, sep=' ', names=['u', 'e', 'a_p', 'inc', 'status']))
df = pd.concat(dl)
df
```

```python
df.to_csv('./earth_trainset01.csv', index=False)
```
